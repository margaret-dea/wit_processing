```python
import pandas as pd
import numpy as np
```

##### r1tvf1v1r

```python
full = pd.read_csv("r1tvf1v1r-fullrun.csv")
partial = pd.read_csv("r1tvf1v1r-partial.csv")
appended = pd.read_csv("r1tvf1v1r-appended.csv")

full['date'] = pd.to_datetime(full['date'])
partial['date'] = pd.to_datetime(partial['date'])
appended['date'] = pd.to_datetime(appended['date'])

print("Full run:")
print("First date:", full['date'].iloc[0])
print("Last date:", full['date'].iloc[-1])

print("\nPartial run:")
print("First date:", partial['date'].iloc[0])
print("Last date:", partial['date'].iloc[-1])

print("\nAppended:")
print("First date:", appended['date'].iloc[0])
print("Last date:", appended['date'].iloc[-1])
```

```python
# Convert 'date' columns to strings
full['date'] = full['date'].astype(str)
partial['date'] = partial['date'].astype(str)
appended['date'] = appended['date'].astype(str)

# Filter partial to only rows up to 2022-04
partial_filtered = partial[partial['date'].str[:7] <= '2022-04']

# Get lists of dates (keep duplicates)
partial_dates = partial_filtered['date'].dropna().tolist()
appended_dates = appended['date'].dropna().tolist()
full_dates = full['date'].dropna().tolist()

# 1. Differences between partial and appended:
# Dates in partial but NOT in appended
partial_not_in_appended = [d for d in partial_dates if d not in appended_dates]

# Dates in appended but NOT in partial
appended_not_in_partial = [d for d in appended_dates if d not in partial_dates]

# 2. Differences between full and combined (partial + appended)
combined_dates = partial_dates + appended_dates

# Dates in full but NOT in combined (missing from both)
full_not_in_combined = [d for d in full_dates if d not in combined_dates]

# Dates in combined but NOT in full (extra in partial+appended)
combined_not_in_full = [d for d in combined_dates if d not in full_dates]

# Print results

print("\nDates in PARTIAL but NOT in APPENDED:")
for d in partial_not_in_appended:
    print(d)

print("\nDates in APPENDED but NOT in PARTIAL:")
for d in appended_not_in_partial:
    print(d)

print("\nDates in FULL but NOT in PARTIAL + APPENDED (missing from both):")
for d in full_not_in_combined:
    print(d)

print("\nDates in PARTIAL + APPENDED but NOT in FULL (extra dates):")
for d in combined_not_in_full:
    print(d)
```
